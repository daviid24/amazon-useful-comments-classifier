### Limpieza y normalizaci√≥n de datos

```python
def normalize_text(text):
    """
    Normaliza el texto: corrige caracteres mal codificados, convierte a min√∫sculas,
    elimina puntuaci√≥n, n√∫meros y stopwords.

    Args:
        text (str): Texto a normalizar.

    Returns:
        str: Texto normalizado.
    """
    # Corregir caracteres mal codificados
    text = unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf-8", "ignore")

    # Convertir a min√∫sculas
    text = text.lower()

    # Eliminar signos de puntuaci√≥n y n√∫meros
    text = re.sub(f"[{re.escape(string.punctuation)}0-9]", " ", text)

    # Tokenizar el texto
    tokens = nltk.word_tokenize(text)

    # Eliminar palabras de parada (stopwords)
    stop_words = set(stopwords.words('spanish'))  # Cambiar idioma si necesario
    tokens = [word for word in tokens if word not in stop_words]

    # Eliminar palabras muy cortas (ej: una letra)
    tokens = [word for word in tokens if len(word) > 1]

    # Unir de nuevo los tokens en una cadena
    normalized_text = " ".join(tokens)

    return normalized_text
```

```python
df['normalized_review_body'] = df['review_body'].apply(normalize_text)
```

```python
df.head()
```

```python
# Funci√≥n para obtener las palabras m√°s frecuentes por categor√≠a
def most_frequent_words_by_category(df, category_column, text_column, top_n=10):
    category_words = {}
    for category, group in df.groupby(category_column):
        # Combinar todos los textos normalizados de la categor√≠a
        all_text = " ".join(group[text_column].dropna())
        # Contar palabras m√°s frecuentes
        most_common = Counter(all_text.split()).most_common(top_n)
        category_words[category] = dict(most_common)
    return category_words

# Obtener palabras m√°s frecuentes por categor√≠a
category_word_counts = most_frequent_words_by_category(df, 'product_category', 'normalized_review_body', top_n=10)

# Graficar palabras m√°s frecuentes por categor√≠a
for category, word_counts in category_word_counts.items():
    plt.figure(figsize=(10, 6))
    plt.bar(word_counts.keys(), word_counts.values())
    plt.title(f"Palabras m√°s frecuentes en la categor√≠a: {category}")
    plt.xlabel("Palabras")
    plt.ylabel("Frecuencia")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Funci√≥n para generar nubes de palabras por puntuaci√≥n
def generate_wordclouds_by_score(df, score_column, text_column):
    for score in sorted(df[score_column].unique()):
        # Filtrar los comentarios para la puntuaci√≥n actual
        text = " ".join(df[df[score_column] == score][text_column].dropna())

        # Crear la nube de palabras
        wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

        # Mostrar la nube de palabras
        plt.figure(figsize=(10, 6))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.title(f"Nube de palabras para la puntuaci√≥n: {score}")
        plt.show()

# Aplicar la funci√≥n usando la columna de texto normalizado y la puntuaci√≥n
generate_wordclouds_by_score(df, score_column='stars', text_column='normalized_review_body')
```

```python
def is_useful_comment(text, score, min_words=10):
    """
    Determina si un comentario es √∫til o no, bas√°ndose en la cantidad de palabras significativas y la puntuaci√≥n.

    Args:
        text (str): Texto del comentario normalizado.
        score (int): Puntuaci√≥n del comentario (1-5).
        min_words (int): N√∫mero m√≠nimo de palabras significativas para considerar √∫til.

    Returns:
        str: 'useful' si el comentario es √∫til, 'not_useful' en caso contrario.
    """
    # Contar palabras significativas (tokens separados por espacio)
    word_count = len(text.split())

    # Determinar si es √∫til: palabras significativas suficientes y puntuaci√≥n extrema
    if word_count >= min_words and score in [1, 5]:
        return 'useful'
    else:
        return 'not_useful'

# Crear la columna 'useful_comment' en el DataFrame
df['useful_comment'] = df.apply(lambda row: is_useful_comment(row['normalized_review_body'], row['stars']), axis=1)

# Verificar resultados
print(df[['review_body', 'normalized_review_body', 'stars', 'useful_comment']].head())
```

## Metodolog√≠a para Definir si un Comentario es √ötil o No

Se utiliz√≥ una combinaci√≥n de criterios para determinar si un comentario es √∫til (`useful`) o no √∫til (`not_useful`). La metodolog√≠a empleada incluye los siguientes pasos:

1. **Preprocesamiento del texto:**
   - Los comentarios fueron normalizados eliminando caracteres mal codificados, puntuaci√≥n, n√∫meros y palabras irrelevantes (stopwords).
   - Esto permiti√≥ extraer solo las palabras significativas de cada comentario.

2. **Definici√≥n de criterios:**
   - Un comentario se consider√≥ **√∫til** si cumpl√≠a con ambas condiciones:
     - **Cantidad de palabras significativas:** El comentario deb√≠a contener al menos 10 palabras significativas, este umbral se seleccion√≥ tras analizar la distribuci√≥n de la longitud de los comentarios y observar que los comentarios con menos de 10 palabras tienden a ser m√°s gen√©ricos y menos informativos. Adem√°s, al revisar ejemplos manualmente, se verific√≥ que 10 palabras son suficientes para expresar una idea √∫til sobre un producto, mientras que comentarios m√°s breves suelen carecer de contexto detallado. despu√©s de la normalizaci√≥n.
     - **Puntuaci√≥n extrema:** El comentario deb√≠a tener una calificaci√≥n de 1 o 5, lo que indica una opini√≥n fuerte (muy positiva o muy negativa).

3. **Creaci√≥n de la etiqueta `useful_comment`:**
   - Los comentarios que cumpl√≠an ambos criterios fueron etiquetados como `'useful'`.
   - El resto fueron etiquetados como `'not_useful'`.

4. **Conversi√≥n a flag binaria:**
   - La etiqueta `useful_comment` se transform√≥ en una variable binaria (`useful_comment_flag`) donde:
     - `1` indica un comentario √∫til.
     - `0` indica un comentario no √∫til.

```python
df['useful_comment_flag'] = df['useful_comment'].apply(lambda x: 1 if x == 'useful' else 0)

# Exportar el DataFrame a un archivo CSV
output_file = "comments_with_usefulness_flag.csv"
df.to_csv(output_file, index=False)

print(f"Archivo exportado exitosamente como: {output_file}")
```

# 2. Construcci√≥n Redes Neuronales

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder, StandardScaler

import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers, optimizers
import seaborn as sns
import nltk
from nltk.corpus import stopwords
```

```python
comments_data = pd.read_csv('comments_with_usefulness_flag.csv', encoding="utf-8")
```

Convierte target en un formato categ√≥rico (one-hot encoding), lo que genera una matriz de etiquetas binarias.

```python
# Preprocesamiento
text_data = comments_data['normalized_review_body']
stars = comments_data['stars']
categories = comments_data['product_category']
target = comments_data['useful_comment_flag']

# Convertir target en categ√≥rica
target = to_categorical(target)
```

- `TfidfVectorizer`: Calcula el peso de cada palabra en relaci√≥n con la importancia dentro de las rese√±as.
- `max_features=10000`: Se limita el vocabulario a las 10000 palabras m√°s frecuentes.
- `fit_transform`: Convierte el texto a una matriz de caracter√≠sticas num√©ricas.
- `LabelEncoder`: Convierte cada categor√≠a de producto en un n√∫mero √∫nico.
- `StandardScaler`: Escala los valores para que tengan media 0 y varianza 1, lo que mejora el rendimiento del modelo.
- `np.hstack`: Combina la matriz TF-IDF, las puntuaciones escaladas y las categor√≠as codificadas en una sola matriz de entrada.

```python
# TF-IDF for text data
tfidf_vectorizer = TfidfVectorizer(max_features=10000)
text_features = tfidf_vectorizer.fit_transform(text_data).toarray()

# Encode product_category as one-hot
category_encoder = LabelEncoder()
categories_encoded = category_encoder.fit_transform(categories)

# Scale stars
scaler = StandardScaler()
stars_scaled = scaler.fit_transform(stars.values.reshape(-1, 1))

# Combine features
import numpy as np
features = np.hstack([text_features, stars_scaled, categories_encoded.reshape(-1, 1)])
```

```python
# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.5, random_state=42)

# Dividir datos de entrenamiento en entrenamiento y validaci√≥n
X_train, X_val, y_train, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

print("Tama√±o del conjunto de validaci√≥n:", X_train.shape)
print("Tama√±o del conjunto de prueba:", X_test.shape)
print("Tama√±o del conjunto de validaci√≥n:", X_val.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l1, l2
import numpy as np

# Definir arquitecturas
def create_model(layers=2, units=128, dropout_rate=0.0, batch_norm=False):
    model = Sequential()
    model.add(Input(shape=(X_test.shape[1],)))
    for _ in range(layers):
        model.add(Dense(units, activation='relu'))
        if batch_norm:
            model.add(BatchNormalization())
        if dropout_rate > 0.0:
            model.add(Dropout(dropout_rate))
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Modelo mixto (Dropout + L1/L2 Regularization)
def create_mixed_model():
    model = Sequential()
    model.add(Input(shape=(X_test.shape[1],)))
    model.add(Dense(128, activation='relu', kernel_regularizer=l1(0.01)))
    model.add(Dropout(0.3))
    model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
    model.add(Dropout(0.3))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Modelos para probar
models = {
    "Basic": create_model(layers=2, units=128, dropout_rate=0.0, batch_norm=False),
    "Deep": create_model(layers=4, units=128, dropout_rate=0.0, batch_norm=False),
    "Dropout": create_model(layers=4, units=128, dropout_rate=0.3, batch_norm=False),
    "BatchNorm": create_model(layers=4, units=128, dropout_rate=0.0, batch_norm=True),
    "Dropout+BatchNorm": create_model(layers=4, units=128, dropout_rate=0.5, batch_norm=True),
    "Mixed (Dropout + L1/L2)": create_mixed_model(),
}

# Entrenamiento y evaluaci√≥n
history = {}
results = {}
for name, model in models.items():
    print(f"Training {name} model...")
    history[name] = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val), verbose=1)
    results[name] = model.evaluate(X_test, y_test, verbose=1)

print(results)
```

```python
def plot_history(history, model_name):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)

    # Graficar precisi√≥n
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'bo', label='Precisi√≥n en entrenamiento')
    plt.plot(epochs, val_acc, 'b', label='Precisi√≥n en validaci√≥n')
    plt.title(f'Precisi√≥n del {model_name}')
    plt.legend()

    # Graficar p√©rdida
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'bo', label='P√©rdida en entrenamiento')
    plt.plot(epochs, val_loss, 'b', label='P√©rdida en validaci√≥n')
    plt.title(f'P√©rdida del {model_name}')
    plt.legend()

    plt.show()

# Graficar para cada modelo
for name, hist in history.items():
    plot_history(hist, name)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay

best_model_name = "BatchNorm"  # Cambia por el modelo que prefieras
best_model = models[best_model_name]

# Ejemplo de datos reales y predicciones (puedes sustituir con tus datos)
y_true = target  # Etiquetas reales
y_pred = best_model.predict(features)  # Predicciones del modelo

# Crear la matriz de confusi√≥n
cm = confusion_matrix(y_true, y_pred)
print("Matriz de Confusi√≥n:\n", cm)

# Calcular las m√©tricas
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='binary')  # Cambia el promedio si hay m√°s de 2 clases
recall = recall_score(y_true, y_pred, average='binary')
f1 = f1_score(y_true, y_pred, average='binary')

# Mostrar las m√©tricas
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# Visualizar la matriz de confusi√≥n
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_true))
disp.plot(cmap="Blues")
```
